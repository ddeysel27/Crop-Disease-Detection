```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")  

model.train(
    data="/content/PlantDoc/plantdoc.yaml",
    epochs=150,
    imgsz=640,
    batch=16,
    lr0=0.0005,     # lower LR improves stability
    optimizer="AdamW",
    device=0,
)
```
