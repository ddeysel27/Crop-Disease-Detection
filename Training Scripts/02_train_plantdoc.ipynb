```python
model = YOLO("yolov8n.pt")

model.train(
    data="processed/PlantDoc/plantdoc.yaml",  # we will create this yaml
    epochs=50,
    imgsz=640,
    batch=16,
    name="yolo_plantdoc",
    project="runs/train"
)
```
